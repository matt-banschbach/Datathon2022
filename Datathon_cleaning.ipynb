```python
import pandas as pd
```

```python
data = pd.read_csv("track.csv")
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data.dropna(subset=['income', 'debt_to_income_ratio', 'census_tract'], inplace=True)
data.loc[data['state_code'].isnull(),'state_code'] = "TX"
```

```python
data.isna().sum()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
data['applicant_age'].unique()
```

```python
age_sort = ['<25','25-34','35-44','45-54','55-64','65-74','>74']
age_enc = OrdinalEncoder(categories=[age_sort])
```

```python
data['applicant_age']= age_enc.fit_transform(data['applicant_age'].values.reshape(-1,1))
```

```python
data.isna().sum()
```

```python
data['applicant_age'].value_counts()
```

```python
data.head()
```

```python
data['debt_to_income_ratio'].unique()
```

```python
dti_sort = ['<20%','20%-<30%','30%-<36%','36%-<41%','41%-<46%','46%-<50%','50%-60%','>60%']
dti_enc = OrdinalEncoder(categories=[dti_sort])
```

```python
data['debt_to_income_ratio']= dti_enc.fit_transform(data['debt_to_income_ratio'].values.reshape(-1,1))
```

```python
data['debt_to_income_ratio'].value_counts()
```

```python
data.isna().sum()
```

```python
data.head()
```

```python
def encode_binary_variable(df, col, val_pos, val_neg):
    df[col] = df[col].replace({val_pos:1, val_neg:0})
```

```python
data['occupancy_type'].unique()
```

```python
encode_binary_variable(data, 'conforming_loan_limit', 'C', 'NC')
encode_binary_variable(data, 'construction_method', 'Site-built', 'Manufactured')
encode_binary_variable(data, 'lien_status', 'First Lien', 'Subordinate Lien')
```

```python
data.head()
data.isna().sum()
```

```python
r_dummies = pd.get_dummies(data['race'])
e_dummies = pd.get_dummies(data['ethnicity'])
s_dummies = pd.get_dummies(data['sex'])
l_dummies = pd.get_dummies(data['loan_type'])
o_dummies = pd.get_dummies(data['occupancy_type'])
d_dummies = pd.get_dummies(data['denial_reason'])
```

```python
d_dummies.head()
```

```python
data = pd.concat([data, r_dummies, e_dummies, s_dummies, l_dummies, o_dummies, d_dummies], axis = 1)
data.head()
data.isna().sum()
```

```python
data.to_csv(r'cleaned_data.csv')
```
